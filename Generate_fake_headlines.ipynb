```python
import random

Subjects=[
"Salman Khan",
"Amit Shah",
"Priyanka Chopra",
"Sachin Tendulkar",
"Donald Trump",
]

activity=[
"was dancing on the street",
"was beaten by monkeys",
"was buying buffalo",
"was drinking soda",
"eating frog",
]

locations=[
"near metro station",
"in Jaipur",
"near kormangla Banglore",
"at India Gate",
"near koklata park",
]

while True:
    sub = random.choice(Subjects)
    loc = random.choice(locations)
    act = random.choice(activity)

    print(f"\n", f"{(sub)} {(act)} {(loc)}")

    user_input = input("Do you want to print another fake headline ? (Yes/No): ").strip().lower()

    if user_input == "no":
        print("I hope you had fun! Good bye")
        break
```
